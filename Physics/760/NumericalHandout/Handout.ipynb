Following is the naive implementation of matrix multiplication

Memory access is generally the limiting facctor in processing speed but it very much depends on the architecture and distribution of python being used.

```python
import numpy as np
import numpy.random as npr
import time
```

```python
def MatMul(A,B):
    C = np.zeros((len(A),len(np.transpose(B))))
    for i in range(len(A)):
        for j in range(len(np.transpose(B))):
            k=0
            while k < len(B):
                C[i,j] += A[i,k]*B[k,j]
                k+=1
    return C

```

```python
n = [10,25,50,75,100,125,150,175,200,250,300]
times_me = []
times_py = []
for i in n:
    A = npr.rand(i,i)
    B = npr.rand(i,i)
    start_me=time.time()
    C = MatMul(A,B)
    end_me=time.time()
    times_me.append(end_me-start_me)
    start=time.time()
    C = np.matmul(A,B)
    end=time.time()
    times_py.append(end-start)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(n,times_me)
plt.xlabel("Dimensions")
plt.ylabel("Time in seconds")
```

```python
plt.plot(n,times_py)
plt.xlabel("Dimensions")
plt.ylabel("Time in seconds")
```

```python
def Riemann(n,s):
    sum = 0
    for i in range(n):
        sum += 1/(i+1)**s
    return sum
```
